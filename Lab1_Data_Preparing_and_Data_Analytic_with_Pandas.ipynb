```python
titanic.sort_values(by=['Pclass'], ascending=True, inplace=False)
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna('30', inplace=True)
titanic['Cabin'].fillna('none', inplace=True)
titanic['Embarked'].fillna('none', inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic.info()
```

```python
titanic['Age'] = titanic['Age'].astype('int')
```

```python
titanic.info()
```

```python
number_survived = len(titanic[titanic['Survived'] == 1])
number_not_survived = len(titanic[titanic['Survived'] == 0])

print('amount of survived {}'.format(number_survived), '\namount of not survived {}'.format(number_not_survived))
```

```python
df_female_temp = titanic[(titanic['Sex'] == 'female') & (titanic['Survived'] == 0)]
df_male_temp = titanic[(titanic['Sex'] == 'male') & (titanic['Survived'] == 0)]

number_female_not_survived = df_female_temp.shape[0]
percentage_female_not_survived = number_female_not_survived * 100 / len(titanic)

number_male_not_survived = df_male_temp.shape[0]
percentage_male_not_survived = number_male_not_survived * 100 / len(titanic)


print('amount of female not survived {}'.format(number_female_not_survived), '\npercentage of female not survived {:.4}'.format(percentage_female_not_survived), '\n')
print('amount of male not survived {}'.format(number_male_not_survived), '\npercentage of male not survived {:.4}'.format(percentage_male_not_survived))
```

```python
df_temp = titanic[titanic['Survived'] == 0]
mode_not_survived_age = df_temp['Age'].mode()
print(mode_not_survived_age)
```

```python
df_temp = titanic[titanic['Survived'] == 0]
average_female_not_survived_age = df_temp[titanic['Sex'] == 'female']['Age'].mean()
average_male_not_survived_age = df_temp[titanic['Sex'] == 'male']['Age'].mean()
print('female: ', average_female_not_survived_age, '\nmale: ', average_male_not_survived_age)
```
